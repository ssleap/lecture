```python
docNouns = {}
maxCount = {}

for file in fileList:
    docid = file[-17:]
    
    with open(file, 'r',encoding='utf-8') as f:
        maxCount[docid] = 0
        content = f.read()
        content = re.sub(r'[\s]{2,}', '',content)
    nouns = {}
    
    for sententce in nltk.sent_tokenize(content):
        
        for word in nltk.word_tokenize(sententce) :
            for w in km().nouns(word) :
                if w in nouns.keys():
                    nouns[w] += 1
                else : nouns[w] = 1
                    
                if maxCount[docid] < nouns[w] :
                    maxCount[docid] = nouns[w]
    docNouns[docid] = nouns
```

```python
for k,v in docNouns.items() :
    print('{0} - {1} - {2}'.format(k, len(v), maxCount[k]))
```

```python
invertedIndex = {}
for noun in uniqueNouns :
    invertedIndex[noun] = []
    for k, v in docNouns.items():
        if noun in v :
            invertedIndex[noun].append(k)
```

```python
print(invertedIndex)
```

```python
kR = 0.5
TF = {}

for k,v in docNouns.items():
    tfList = {}
    for w in v :
        tfList[w] = kR + (1-kR)*(v[w] / maxCount[k])
        print('{5} | {0} = {1} + {2} * ({3}/{4})'.format(tfList[w], kR, 1-kR, v[w], maxCount[k], w))
    TF[k] = tfList
```

```python
from math import log10

docSize = len(fileList)
TFIDF = {}


for k,v in TF.items():
    idfList = {}
    for t in v:
        try :
            idf = log10(docSize / len(invertedIndex[t]))
            idfList[t] = v[t] * idf
            print('{0} | {1} * log({2} / {3}) = {4}'.format(
                t,v[t],docSize,idf,idfList[t]
            ))
        except :
            print('god')
        
    TFIDF[k] = idfList
    
        
```

```python
query = '미국과 중국 중 미국 이겨라!'

queryNouns = {}
maxCount = 0


        
for word in nltk.word_tokenize(query) :
    for w in km().nouns(word) :
        if w in queryNouns.keys():
            queryNouns[w] += 1
        else : 
            queryNouns[w] = 1
                    
        if maxCount < queryNouns[w] :
            maxCount = queryNouns[w]
print(len(queryNouns), maxCount)
print(queryNouns)
```

```python
queryWeight = {}

for k,v in queryNouns.items() :
    queryWeight[k] = kR + (1- kR) * (v/maxCount)
    print('{5} | {0} = {1} + {2} * ({3}/{4})'.format(kR, 1-kR, v, maxCount, queryWeight[k], k))
```

```python
from math import sqrt

docLength = {} # 문서1 : 길이, 문서2 : 길이 ...

for k, v in TFIDF.items() :
    sumPow = 0.0
    
    for t in v :
        sumPow += v[t]**2
        
    docLength[k] = sqrt(sumPow)
    
#     k # 문서 ID
#     v # 단어1 : 가중치, 단어2 :가중치...
    
    
    
```
