# 110

```python
def per(num):
    if num == 1: return True
    
    pro_div_sum = 0
    for i in range(1, num):
        if num % i == 0:
            pro_div_sum += i
    if pro_div_sum == num: 
        return True
    else:
        return False

for i in range(1,10001):
    print('Is', i , 'perfect? :',per(i))
```

```python
def check_ana(s,t):
    if len(s) != len(t):
        return False
    a = list(s)
    b = list(t)
    for i in range(len(a)):
        if a[i] in b :
            for j in range(len(b)) :
                if b[j] == a[i]:
                    del b[j]
        else :
            return False
    return True
        
a = input()
b = input()
print(check_ana(a,b))
    
    
```
